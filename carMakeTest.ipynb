```python
player = "Suley"
```

```python
base_path = Path(r"C:\Projekty\Coding\Python\F1PredsPPT\CarBase")
img_names = set(["Axles", "Core", "DriverSeat", "FrontWing", "RearWing", "Wheels"])

img_paths = {name: base_path / f"{name}.png" for name in img_names}
```

```python
img_colours = {name: (100, 100, 100) for name in img_names}

img_colours["Wheels"] = (0,0,0)
img_colours["Core"] = PLAYERS_TO_COLOURS[player]
img_colours["RearWing"] = tuple((int(0.8 * i) for i in PLAYERS_TO_COLOURS[player]))
# img_colours["RearWing"] = img_colours["FrontWing"]
img_colours["FrontWing"] = img_colours["Core"]
img_colours["DriverSeat"] = img_colours["RearWing"]
```

```python
img_colours
```

```python
# %matplotlib tk
```

```python
tot_image = Image.new("RGBA", (54, 21), (255, 255, 255, 0))
```

```python
# tot_img = np.zeros((21, 54, 3))

for img_name in img_names:
    image = Image.open(img_paths[img_name])
    image_colour = Image.new("RGB", (54, 21), img_colours[img_name])
    tot_image = Image.composite(image_colour, tot_image, image)
```

```python
plt.imshow(tot_image)
```

```python
for name in PLAYERS_TO_COLOURS:

    img_colours = {name: (100, 100, 100) for name in img_names}

    img_colours["Wheels"] = (0,0,0)
    img_colours["Core"] = PLAYERS_TO_COLOURS[name]
    img_colours["RearWing"] = tuple((int(0.8 * i) for i in PLAYERS_TO_COLOURS[name]))
    img_colours["FrontWing"] = img_colours["Core"]
    img_colours["DriverSeat"] = img_colours["RearWing"]

    tot_image = Image.new("RGBA", (54, 21), (255, 255, 255, 0))

    for img_name in img_names:
        image = Image.open(img_paths[img_name])
        image_colour = Image.new("RGB", (54, 21), img_colours[img_name])
        tot_image = Image.composite(image_colour, tot_image, image)

    tot_image.save(f"Cars2024/{name}.png")
```
